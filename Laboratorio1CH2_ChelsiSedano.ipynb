En el presente trabajo, se pretende realizar dos estrategias de inversión una pasiva y una activa, esto con la finalidad de observar cuál será la mejor opción para el millón de pesos que pone en juego el inversionista en curso. 

La inversión pasiva evita el análisis de acciones y transacciones individuales dentro y fuera del mercado. El objetivo de los inversores pasivos es obtener rentabilidades indexadas. No buscan activamente superar el rendimiento del mercado, sino que intentan replicar el rendimiento del índice establecido.(Por ejemplo, podrías adquirir un fondo que replique el comportamiento del IBEX 35). Para este proyecto, se está haciendo uso del ETF IShare NAFTRAC.

La inversión activa es un método de inversión financiera que se caracteriza por adoptar un enfoque pragmático de participación en el mercado bursátil y buscar superar el rendimiento medio del mercado de valores.

# <font color= #6B6B6B> 4. Data Description </font>

Se descarga los archivos de NAFTRAC que se encuentra en la carpeta de files, después se ordenan por fecha , se eliminan todos aquellos activos que no se necesitan y se cambia el nombre de algunos activos porque ahora para buscarlos en Yahoo finance se llaman diferente. 

Después se utiliza la paquetería de Yahoo Finance para bajar los datos y así obtener los cierres.

```python
#Variable que guarda la direccion de los archivos "files"
abspath = path.abspath('files')

# Crear objeto
mis_datos = Acomodar_Data(abspath)

## Acomodar datos
file_dates = mis_datos.obtener_fechas()
data_files = mis_datos.correcciones_arch(file_dates)
tickers = mis_datos.tickers_unq(data_files,file_dates)
dates = mis_datos.formato_fechas(file_dates)
yahoo_data = mis_datos.info_yahoo(dates,tickers)
data_close = mis_datos.obtener_cierres(tickers,yahoo_data)
data_closeT= mis_datos.closes_fnec(data_close,dates)
```

## INVERSION PASIVA

Se obtiene la base con el capital, el rendimiento y el rendimiento acumulado total de una inversion pasiva.

```python
#Crear objeto
mi_inv_pasiva = Inversion_pasiva()

capital = 1000000
comision = 0.00125
fir_data= mi_inv_pasiva.creacion_tabla(data_files, file_dates,data_closeT,capital,comision)

# Efectivo 
cash = (1-fir_data['Peso (%)'].sum())* capital

# Sumatoria de las columnas de las comisiones
comision_sum = fir_data['Comisiones'].sum()

# Obteniendo inversion pasiva
passive_inv = mi_inv_pasiva.inv_fechas(dates,fir_data,data_closeT,file_dates,cash,capital)
inv_passive= mi_inv_pasiva.base_inv_pasiva(passive_inv)
```

```python
inv_passive
```

Se separa por las fechas que fueron antes de la pandemia. 

```python
# PRE PANDEMIC
prepa = inv_passive.loc[0:25]
prepa
```

Se obtiene las que fueron en pos pandemia 

```python
# POS PANDEMIC
pospa= inv_passive.loc[25:].copy()
pospa.loc[25, 'Capital'] = capital
pospa.loc[25, 'Rendimiento']= 0
pospa['Rendimiento Acumulado'] = pospa['Rendimiento'].cumsum()
pospa
```

## INVERSION ACTIVA

Se fijan las constantes como los activos, los precios de los activos, la fecha donde empieza, la tasa libre de riesgo y la  periodicidad que en este caso es mensual.

```python
activ_tick= fir_data['Ticker'].tolist()
j_prices= data_closeT[activ_tick]
fecha_act= '2020-02-28'
tasa_lr= 0.0425
periodicidad= 12
```

Se obtiene la base con el capital, el rendimiento y el rendimiento acumulado total de una inversion pasiva.

```python
# Crear objeto
mi_inv_activa = Inversion_activa()

prices_pre = mi_inv_activa.prices_pre(j_prices,fecha_act)

# Seleccionar el portafolio con mayor sharpe
mat_cov = prices_pre.cov() * np.sqrt(periodicidad)
portafolios= mi_inv_activa.portafolio_act(prices_pre, periodicidad,tasa_lr,mat_cov)
w_opt = mi_inv_activa.obtener_w_opt(portafolios)

net_capital = 1000000 - cash

pos_prices_f,lista= mi_inv_activa.pos_prices_f(j_prices,fecha_act)
pos_prices,lista2= mi_inv_activa.pos_prices(j_prices,fecha_act)

port_opt= mi_inv_activa.obtener_portafolio_optimo(pos_prices_f,w_opt, net_capital,comision)

# Restar comisiones 
cash_act = cash - port_opt['Comisiones'].sum()

## Seleccionar up y down
down,up = mi_inv_activa.updown(pos_prices)

new_portfolio= mi_inv_activa.obtener_nuevo_port(j_prices,port_opt,down,comision)

valor_portafolio, acciones_compra, acciones_venta, comisiones_compra,comisiones_venta = mi_inv_activa.valor_portafolio(cash,pos_prices_f,j_prices,port_opt,down,up,comision)

df_activa= mi_inv_activa.df_activa(valor_portafolio, lista)
df_activa
```

Se muestra el histotico de operaciones que se fueron realizando a lo largo de la inversion activa.

```python
df_operaciones= mi_inv_activa.df_operaciones(lista2,acciones_compra,acciones_venta,comisiones_compra,comisiones_venta)
df_operaciones
```

## Mis medidas de desempeño

```python
mis_medidas_desempeno= mi_inv_activa.medidas_desempeno(pospa,df_activa,periodicidad)
mis_medidas_desempeno
```

# <font color= #6B6B6B> 5. Conclusiones </font>

Como  se mencionaba anteriormente, la inversion pasiva lo que busca es unicamente replicar lo que hace un ETF, entonces las personas que hacen este tipo de inversiones lo que quiere lograr es tener "segura" su inversion que finalmente a la larga va a lograr tener un rendimiento pero las personas que hacen inversiones activas tienen "más riesgo", pero como dicen "a mayor riesgo, mayor ganancia". Las personas que hacen inversiones activas estan constantemente comprando y vendiendo. Creo que esta estrategia es muy buena, con comisiones relativamente bajas la rentabilidad puede ser mejor pero hay que estar revisando que las comisiones no se lleven todo el rendimiento pues sabemos que en la vida real las comisiones suelen ser más elevadas.

En esta ocasion, en la inversión pasiva total tuvimos un rendimiento mínimo en casi todo el tiempo tuvimos perdida hasta el final. Por otro lado, separandolo por invirtiendo antes y después de la pandemia pordemos observar que si hubieramos invertido antes de la pandemia hubieramos tenido perdidas pero si hubieramos empezado durante la pandemia hubieramos tenido una ganancia mayor a alguna de las anteriores opciones. Esto puede deberse a diferentes factores pero uno que suena lógico es que durante la pandemia hubo un "boom" por invertir desde casa, ya que se incitaba a todo tipo de peronas a invertir aunque fuera con muy poco y al estar desde casa sin tener mucho que hacer las personas se animaron a intentar invertir.

Por último, vemos la inversion activa que nos da un rendimiento mayor al de pasiva, como ambos se analizan en el mismo periodo de tiempo se observa que lo más conveniente para un inversior que hubiera querido invertir desde el 2020 era esperarse a la llegada de la pandemia para tener mejores rendimientos.
